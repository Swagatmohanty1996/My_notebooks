```python
sym=['^GSPC']

start_date='2018-01-01'
end_date='2019-07-25'

prices_df=data.DataReader(sym,'yahoo',start_date,end_date)
```

```python
prices_df.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, ax= plt.subplots(figsize=(20,10))
ax.plot(prices_df.index, prices_df['Adj Close'])
plt.title('S&P 500')
```

```python
fig, ax= plt.subplots(2,1,figsize=(20,10))

ax[0].plot(prices_df.index, prices_df['Adj Close'])
ax[1].plot(prices_df.index, prices_df['Volume'])
plt.title('S&P 500')
```
